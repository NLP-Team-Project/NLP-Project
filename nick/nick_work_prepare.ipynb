```python
def acquire_repos():
    repos = open('repos.txt', 'r')
    repo_list = repos.readlines()
    repositories = []
    for l in repo_list:
        repositories.append(re.findall(r'\w[^\']+', l)[0])
    repos.close()
    repositories.remove('majdsassi/Sport-IPTV-FR')
    repositories.remove('ssgerson/lista.m3u')
    repositories.remove('Agoengwew/wew')
    return repositories

acquire_repos()[:5]
```
