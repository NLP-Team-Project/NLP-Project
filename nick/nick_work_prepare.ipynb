```python
import re
import pandas as pd

import prepare as prp
```

```python
def acquire_repos():
    repos = open('repos.txt', 'r')
    repo_list = repos.readlines()
    repositories = []
    for l in repo_list:
        repositories.append(re.findall(r'\w[^\']+', l)[0])
    repos.close()
    repositories.remove('majdsassi/Sport-IPTV-FR')
    repositories.remove('ssgerson/lista.m3u')
    repositories.remove('Agoengwew/wew')
    repositories.remove('sugeth/xxx')
    return repositories

repos = acquire_repos()
len(repos)
```

```python
repositories = prp.scrape_github_data(repos)
```

```python
repositories[0]
```

```python
repo_name = []
language = []
readme = []
```

```python
for dictionary in repositories:
    repo_name.append(dictionary['repo'])
    language.append(dictionary['language'])
    readme.append(dictionary['readme_contents'])
```

```python
all_repos = {'repo':repo_name,
             'language':language,
             'readme_contents':readme}
```

```python
all_repos = pd.DataFrame(all_repos)
```

```python
all_repos.to_csv('all_repos.csv', )
```
